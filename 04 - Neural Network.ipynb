```python
# Importing the saved model is used from directory
clf = joblib.load('NN_classifier.sav')
```

```python
prediction = clf.predict(X_test)
```

```python
acc = confusion_matrix(y_test, prediction)
```

```python
print('confusion matrix: \n', acc)
```

```python
def accuracy(confusion_matrix):
    diagonal = confusion_matrix.trace()
    elements = confusion_matrix.sum()
    return diagonal / elements
```

```python
print('accuracy: ', accuracy(acc))
```

```python
def transform_image(image):
    for pixel in range(len(image)):
        image[pixel] = (255 - image[pixel][0]) / 256
    return np.array(image)
```

```python
actuals = list()
predictions = list()

for i in range(10):
    actuals.append(i)
    filename = 'NN_images/' + str(i) + '.png'
    img = Image.open(filename)
    data = list(img.getdata())
    data = transform_image(data)
    prediction = clf.predict([data])
    predictions.append(prediction)
    print('input: ', i, '\t prediction: ', prediction)
```

```python
acc = confusion_matrix(actuals, predictions)
print(acc)
acc.trace()
```

```python
def accuracy(cm):
    diagonal = cm.trace() # true positives
    elements = cm.sum() # num predictions
    return diagonal / elements
```

```python
print('accuracy: ', accuracy(acc))
```
